### Treinamento e Avaliação de algoritmos de Machine Learning

#### Rede Neura Sequencial

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
dados = pd.read_csv('Consumo_cerveja_2.csv')
print(dados)
```

```python
dados.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    dados[['Temperatura Media (C)', 'Temperatura Minima (C)', 'Temperatura Maxima (C)', 'Precipitacao (mm)', 'Final de Semana']],
    dados[['Consumo de cerveja (litros)']],
    test_size=0.2,
    random_state=86
)
```

```python
modelo_rede = tf.keras.Sequential([
    tf.keras.layers.Dense(102, activation='relu', input_shape=(5,)),
    tf.keras.layers.Dense(86, activation='relu'),
    tf.keras.layers.Dense(94, activation='relu'),
    tf.keras.layers.Dense(63, activation='relu'),
    tf.keras.layers.Dense(23, activation='relu'),
    tf.keras.layers.Dense(9, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
modelo_rede.summary()
```

```python
modelo_rede.compile(optimizer='adam', loss='mse')
```

```python
modelo_rede.fit(x_train, y_train, epochs=2100, batch_size=86, verbose=1)
```

```python
previsoes_rede = modelo_rede.predict(x_test)
```

```python
MAE_rede_neural = mean_absolute_error(y_test, previsoes_rede)
MSE_rede_neural = mean_squared_error(y_test, previsoes_rede)
R2_rede_neural = r2_score(y_test, previsoes_rede)
```

```python
print(f'Erro médio Absoluto Algoritmo Rede Neural: {MAE_rede_neural}')
print(f'Erro médio Quadrático Algoritmo Rede Neural: {MSE_rede_neural}')
print(f'Coeficiente de  Determinação Algoritmo Rede Neural: {R2_rede_neural}')
```
